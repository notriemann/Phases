```python
import numpy as np
import time as timer
```

## Make a sample task with a loop for speed-up reference

```python
#take the current time
timer2 = timer.time()


#define function to do the task and take 0.1 seconds of delay
def work(i):
    timer.sleep(0.1)
    return i

#do 100 iterations
N=100


#sum all of the integers to N
an = 0
for i in range(N):
    
    an = an + work(i)
    
print("Sum = ", an)
print("Time ", timer.time() - timer2)
    
```

## Parallelization example with joblib

```python
from joblib import Parallel, delayed
```

```python
#use 4 cores for ~4 times speedup
N_cores = 4


timer2 = timer.time()

results = Parallel(n_jobs=4)(
    delayed(work)(i) for i in range(N)
)

N_sum = sum(results)

print("Sum = ", N_sum)
print("Time ", timer.time() - timer2)
```

## Parallelization example with multiprocessing

```python
from multiprocessing import Pool
```

```python
N_cores = 4

timer2 = timer.time()

with Pool(N_cores) as p:
    results = p.map(work, range(N))

N_sum = sum(results)

print("Sum = ", N_sum)
print("Time ", timer.time() - timer2)
```
